```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('Chinook_Sqlite.sqlite')  # Use the exact name of your uploaded file
tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table';", conn)
print("Available tables:")
print(tables)
```

```python
query = """
SELECT C.FirstName || ' ' || C.LastName AS Customer,
       SUM(I.Total) AS TotalSpent
FROM Customer C
JOIN Invoice I ON C.CustomerId = I.CustomerId
GROUP BY Customer
ORDER BY TotalSpent DESC
LIMIT 5;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT G.Name AS Genre, SUM(IL.Quantity) AS ItemsSold
FROM InvoiceLine IL
JOIN Track T ON IL.TrackId = T.TrackId
JOIN Genre G ON T.GenreId = G.GenreId
GROUP BY Genre
ORDER BY ItemsSold DESC;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT BillingCountry, ROUND(AVG(Total), 2) AS AvgInvoice
FROM Invoice
GROUP BY BillingCountry
ORDER BY AvgInvoice DESC;
"""
pd.read_sql(query, conn)
```

```python
query = """
CREATE VIEW IF NOT EXISTS EmployeeSales AS
SELECT E.FirstName || ' ' || E.LastName AS Employee,
       COUNT(I.InvoiceId) AS TotalInvoices,
       SUM(I.Total) AS TotalSales
FROM Employee E
JOIN Customer C ON E.EmployeeId = C.SupportRepId
JOIN Invoice I ON C.CustomerId = I.CustomerId
GROUP BY Employee;
"""
conn.execute(query)
```

```python
pd.read_sql("SELECT * FROM EmployeeSales;", conn)
```

```python
query = """
SELECT FirstName || ' ' || LastName AS Customer, Total
FROM (
    SELECT C.FirstName, C.LastName, SUM(I.Total) AS Total
    FROM Customer C
    JOIN Invoice I ON C.CustomerId = I.CustomerId
    GROUP BY C.CustomerId
) AS Totals
WHERE Total > (SELECT AVG(Total) FROM Invoice);
"""
pd.read_sql(query, conn)
```
